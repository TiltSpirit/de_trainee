**11.Сгруппируйте людей по возрастным группам *young*, *adult*, *retiree*, где:**
* *young* соответствует 16-35 лет
* *adult* - 35-70 лет
* *retiree* - 70-100 лет

**Проставьте название соответсвтуещей группы для каждого человека в новой колонке AgeGroup**

```python
df_age_group = df.copy()
df_age_group.head(2)
```

```python
def age_group_sort(x):
    if x >= 16 and x < 35:
        return 'young'
    elif x >= 35 and x < 70:
        return 'adult'
    elif  x >= 70 and x < 100:
        return 'retiree'
    else:
        return 'no info given'
df_age_group['AgeGroup'] = df_age_group['age'].apply(age_group_sort)
df_age_group
```

**12-13. Определите количество зарабатывающих >50K в каждой из возрастных групп (колонка AgeGroup), а также выведите название возрастной группы, в которой чаще зарабатывают больше 50К (>50K)**

```python
age_group_working = df_age_group.groupby(['AgeGroup', 'salary'])['age'].agg([('amount', 'count')])
print(age_group_working)
age_bi = age_group_working.reset_index() #bi - better income, same as above
result_group = age_bi[(age_bi['salary'] == ">50K")].sort_values('amount', ascending=False).iloc[0]
result_group
print(f'\nThe oldest age group who earned more than 50K is "{result_group['AgeGroup']}" and there were {result_group['amount']} of them')
```

**14. Сгруппируйте людей по типу занятости (колонка occupation) и определите количество людей в каждой группе. После чего напишите функциюю фильтрации filter_func, которая будет возвращать только те группы, в которых средний возраст (колонка age) не больше 40 и в которых все работники отрабатывают более 5 часов в неделю (колонка hours-per-week)**

```python
grouped_oc = df_age_group.iloc[:, 0:13:6]
grouped_oc
```

```python
table_oc = df_age_group.groupby('occupation').agg({'age': ['count', 'mean'], 'hours-per-week': ['min']}).reset_index()
table_oc.columns = ['occupation', 'number_of_people', 'avg_age', 'min_worktime']
table_oc
```

function making in the nutshell

```python
def filter_occupation(some_table):
    final_occupation = []
    for i in range(len(some_table)):
        x = some_table.iloc[i]
        if x['avg_age'] <= 40 and x['min_worktime'] > 5:
            final_occupation.append(x)
    if final_occupation:
        return pd.DataFrame(final_occupation)
    else:
        print('Nothing is matching the filtering conditions')
        return None
```

```python
drumroll = filter_occupation(table_oc)
drumroll
```
