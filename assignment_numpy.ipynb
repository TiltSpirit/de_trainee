* __Задача 1__: Подсчитать произведение ненулевых элементов на диагонали прямоугольной матрицы.  
 Например, для X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]) ответ 3.

```python
X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]])
checker = np.full((5,3), 0)
min(len(X), len(X[0])) # использую для определения максимальной длины диагонали матрицы 
```

python only

```python
def diagonal_multiply(matrix):
    starting_point = 1
    zero_counter = 0
    for i in range(min(len(matrix), len(matrix[0]))):
        element = matrix[i][i]
        if element != 0:
            starting_point *= element
        else:
            zero_counter += 1
            continue
    if zero_counter == min(len(matrix), len(matrix[0])):
        print('unlucky bro, all diagonal elements are 0')
        return None
    else:
        return starting_point
    
```

```python
diagonal_multiply(checker)
diagonal_multiply(X)
```

using numpy

```python
def numpy_diagonal_multiply(matrix):
    diagonal = np.diag(matrix)
    product = np.prod(diagonal[diagonal != 0])
    if np.any(diagonal) == False:
        print('unlucky bro, all diagonal elements are 0')
        return None
    return product
```

```python
numpy_diagonal_multiply(checker)
numpy_diagonal_multiply(X)
```

* __Задача 2__: Даны два вектора x и y. Проверить, задают ли они одно и то же мультимножество.  
  Например, для x = np.array([1, 2, 2, 4]), y = np.array([4, 2, 1, 2]) ответ True.

```python
#python one
x = np.array([1, 2, 2, 4])
y = np.array([2, 4, 1, 2])
```

```python
def multiset_checker(vector_1, vector_2):
    x_cnt = {}
    for i in vector_1:
        x_cnt[i] = x_cnt.get(i, 0) + 1
    y_cnt = {}
    for i in vector_2:
        y_cnt[i] = y_cnt.get(i, 0) + 1
    return x_cnt == y_cnt
```

```python
multiset_checker(x, y)
```

```python
#numpy one
def numpy_multiset_checker(vector_1, vector_2):
    return np.array_equal(np.sort(vector_1), np.sort(vector_2))
```

```python
numpy_multiset_checker(x, y)
```

* __Задача 3__: Найти максимальный элемент в векторе x среди элементов, перед которыми стоит ноль. 
 Например, для x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0]) ответ 5.

```python
x = np.array([0, 9, 0, 3, 0, 0, 5, 7, 0])
y = np.array([0, -3, 6])
```

```python
#python one
def max_finder(arr):
    max = None
    for i in range(1, len(arr)):
        if arr[i - 1] == 0:
            if (max is None) or arr[i] > max:
                max = arr[i]
    return max
```

```python
max_finder(y)
```

```python
#numpy one
def numpy_max_finder(arr):
    zero_idx = np.where(arr[:-1] == 0)[0]
    numbers_after_idx = zero_idx + 1
    numbers = arr[numbers_after_idx]
    return np.max(numbers)
```

```python
numpy_max_finder(x)
```

* __Задача 4__: Реализовать кодирование длин серий (Run-length encoding). Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить.  
 Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) ответ (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
x = np.array([2, 2, 4, 3, 2, 3, 3])
```

```python
#python one
def RLE(arr):
    values = []
    streak = []
    count = 1
    for i in range(1, len(arr)):
        if arr[i] == arr[i-1]:
            count += 1
        else:
            values.append(arr[i-1])
            streak.append(count)
            count = 1
    values.append(arr[-1])
    streak.append(count)
    return list(map(int, values)), list(map(int, streak))
```

```python
print(f'array is {x}\n\n encoded is: {RLE(x)}')
```

* __Задача 5__: Даны две выборки объектов - X и Y. Вычислить матрицу евклидовых расстояний между объектами. Сравните с функцией scipy.spatial.distance.cdist по скорости работы.

```python
X = np.array([[0], [11]])
Y = np.array([[3], [-2]])
```

```python
from scipy.spatial.distance import cdist
import time
```

```python
'''def distance(X, Y):
    return np.sqrt(np.sum(np.square(X - Y)))'''
```

```python
def distance(X, Y):
    n = X.shape[0]
    m = Y.shape[0]
    matrix = np.zeros((n, m))
    for i in range(n):
        for j in range(m):
            diff = X[i] - Y[j]
            matrix[i, j] = np.sqrt(np.sum(diff**2))
    return matrix
```

```python
np_time_s = time.time()
d1 = distance(X, Y)
np_time_f = time.time() - np_time_s
print(f'{d1} and np time is {np_time_f} sec\n\n')

sci_time_s = time.time()
d2 = cdist(X, Y, 'euclidean')
sci_time_f = time.time() - sci_time_s
z = (np_time_f/sci_time_f) * 100
print(f'{d2} and sci time is {sci_time_f} sec\n\n np / sci relation is {z:.3f}%')
```
